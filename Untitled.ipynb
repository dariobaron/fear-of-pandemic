```python
fig, ax = plt.subplots(figsize=(5,3))

xs = np.linspace(0,1,1001)

ax.axvline(0.5, color="black", label="Fixed")
ax.plot(xs, beta.pdf(xs,5,5), color="green", label="Unimodal")
ax.plot(xs, beta.pdf(xs,5/2,5/4,scale=3/4), color="blue", label="Asy. high")
ax.plot(xs, beta.pdf(xs,5/4,5/2,loc=1/4,scale=3/4), color="red", label="Asy. low")
ax.plot(xs, 0.5*beta.pdf(xs,8,2)+0.5*beta.pdf(xs,2,8), color="purple", label="Bimodal")

ax.legend()
ax.spines[["top","right"]].set_visible(False)
ax.set_xticks(np.linspace(0,1,3))
ax.set_yticks(np.linspace(0,3,2))

fig.tight_layout()
fig.savefig("plots/fear-distributions.png")
```
