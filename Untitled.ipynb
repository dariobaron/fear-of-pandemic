```python
def get_lists_SIR(nodes_dict):
    SIR_dict = {"S":[], "I":[], "R":[]}
    for node,status in nodes_dict.items():
        SIR_dict[status].append(node)
    return SIR_dict

def SIR_net(G, beta, mu, initial_infecteds):

    '''
    INPUT 
    G : a network
    beta : transmission per link
    mu : recovery rate
    initial_infecteds : node or iterable of nodes that will be infected at the beginning 
    
    OUTPUT
    t,S,I,R numpy arrays with series of time steps, and series of number of infected, susceptible, recovered
    '''

    '''
    G.number_of_nodes() : to get the number of nodes
    nx.set_node_attributes(G,dictionary_node_attribute, 'attribute_name') : nodes have as attribute attribute_name
    G.node[i]['attribute']= attibute_value : set attribute of node i to attribute_value
    nx.nodes(G) : iterable on nodes
    nx.all_neighbors(G, i) : iterable on nodes neighbors of i
    '''
    
    # count the nodes
    N = G.number_of_nodes()
    
    # setting all nodes to suscettible state at the beginning
    nx.set_node_attributes(G, "S", "status")
    # setting the initial infected nodes
    for i in initial_infecteds:
        nx.set_node_attributes(G, {i : {"status": "I"}})
    # setting the initial arrays for S, I, R populations
    S = np.array([N - len(initial_infecteds)])
    I = np.array([len(initial_infecteds)])
    R = np.array([0])
    # setting the lists of nodes in each status
    SIR_dict = get_lists_SIR(nx.get_node_attributes(G, "status"))
    S_nodes = SIR_dict["S"]
    I_nodes = SIR_dict["I"]
    R_nodes = SIR_dict["R"]
    
    # loop over the time-steps until there is no infectious
    t = [0]
    while(I[-1] > 0):
        
        # infecting
        new_I_nodes = np.array([])
        for i in I_nodes:
            S_neighbours = np.array([neighbour for neighbour in nx.all_neighbors(G, i) if G.nodes[neighbour]["status"] == "S"])
            gets_infected = np.random.binomial(1, beta, size=len(S_neighbours)).astype(bool)
            new_I_nodes = np.append(new_I_nodes, S_neighbours[gets_infected])
        # network nodes update
        nx.set_node_attributes(G, {node:{"status":"I"} for node in new_I_nodes})
            
        #recovering
        gets_recovered = np.random.binomial(1, mu, size=I[-1]).astype(bool)
        new_R_nodes = np.array(I_nodes)[gets_recovered].tolist()
        # network nodes update
        nx.set_node_attributes(G, {node:{"status":"R"} for node in new_R_nodes})
        
        # lists update
        SIR_dict = get_lists_SIR(nx.get_node_attributes(G, "status"))
        S_nodes = SIR_dict["S"]
        I_nodes = SIR_dict["I"]
        R_nodes = SIR_dict["R"]
        
        # populations update
        S = np.append(S, len(S_nodes))
        I = np.append(I, len(I_nodes))
        R = np.append(R, len(R_nodes))
        
        # time update
        t.append(t[-1]+3/24)
    
    return t, S, I, R
```

```python
BAG= nx.read_adjlist("input/barabasi-albert_10000_4.adjlist", nodetype=int)
```

```python
beta= 1/2*3/24/4
mu= 1/6*3/24 # recovery probability

t,S,I,R = SIR_net(BAG, beta, mu, np.random.choice(np.arange(N), size=2, replace=False))
t = np.array(t)
```

```python
plt.plot(t, S, label="det S")
plt.plot(t, I, label="det I")
plt.plot(t, R, label="det R")
plt.plot(data["S"], linestyle="--", label="stoch S")
plt.plot(data["E"], linestyle="--", label="stoch E")
plt.plot(data["I"], linestyle="--", label="stoch I")
plt.plot(data["R"], linestyle="--", label="stoch R")
plt.legend()
#plt.xlim([0,30])
#plt.ylim([0,500])
plt.grid()
```
