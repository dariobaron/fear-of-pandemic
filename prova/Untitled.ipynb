```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
G = nx.barabasi_albert_graph(1000, 4)
```

```python
len(G.nodes())
```

```python
nx.draw(G)
```

```python
adj_mat = nx.to_pandas_adjacency(G)
adj_mat
```

```python
ax = adj_mat.sum().hist(bins=np.logspace(np.log10(4),np.log10(adj_mat.sum().max()),15))
ax.set_xscale("log")
ax.set_yscale("log")
```

```python
nx.write_adjlist(G, "adjlist.dat")
```

```python
nx.write_weighted_edgelist(G, "weighted_edgelist.dat")
```
